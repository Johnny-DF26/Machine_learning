```python
df.columns = ['ID', 'Idade', 'Experiencia', 'Renda', 'ZIP_code', 'Familia', 'Media_mensao_CC', 'Educacao',
             'Hipoteca', 'Emprestimo_pessoal', 'Contas_aux', 'Conta_CD', 'Online', 'Cartao_credito']
```

```python
sns.set(style='ticks')
# Definir um mapeamento de cores para a variÃ¡vel "Personal Loan"
cores = {0: 'blue', 1: 'orange'}

# DistribuiÃ§Ã£o das variÃ¡veis numÃ©ricas em relaÃ§Ã£o a "Personal Loan"
num_vars = ["Idade", "Experiencia", "Renda", "Media_mensao_CC", "Hipoteca"]

for var in num_vars:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=var, hue="Emprestimo_pessoal", element="step", palette=cores)
    plt.title(f"DistribuiÃ§Ã£o de {var} em relaÃ§Ã£o a Emprestimo Pessoal")
    plt.legend(title="Emprestimo", labels=[1, 0])
    plt.show()
    
# Quantidades das variÃ¡veis categÃ³ricas em relaÃ§Ã£o a "Personal Loan"
cat_vars = ["Familia", "Educacao", "Contas_aux", "Conta_CD", "Online", "Cartao_credito"]

for var in cat_vars:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=var, hue="Emprestimo_pessoal")
    plt.title(f"Quantidades de {var} em relaÃ§Ã£o a Emprestimo Pessoal")
    plt.legend(title="Emprestimo", labels=[1, 0])
    plt.show()
```

```python
# Definir um mapeamento de cores para a variÃ¡vel "Personal Loan"
cores = {0: 'blue', 1: 'orange'}

# Configurar o layout dos subplots
plt.figure(figsize=(15, 10))
plt.subplots_adjust(hspace=0.5)

# DistribuiÃ§Ã£o das variÃ¡veis numÃ©ricas em relaÃ§Ã£o a "Personal Loan"
num_vars = ["Idade", "Experiencia", "Renda", "Media_mensao_CC", "Hipoteca"]

for i, var in enumerate(num_vars, start=1):
    plt.subplot(2, 3, i)
    sns.histplot(data=df, x=var, hue="Emprestimo_pessoal", element="step", palette=cores)
    plt.title(f"DistribuiÃ§Ã£o de {var}")
    plt.legend(title="Emprestimo", labels=["NÃ£o Aceitou", "Aceitou"])
plt.suptitle("AnÃ¡lise em relaÃ§Ã£o a Emprestimo pessoal", fontsize=16)

plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplots_adjust(hspace=0.5)

# Quantidades das variÃ¡veis categÃ³ricas em relaÃ§Ã£o a "Personal Loan"
cat_vars = ["Familia", "Educacao", "Contas_aux", "Conta_CD", "Online", "Cartao_credito"]

for i, var in enumerate(cat_vars, start=1):
    plt.subplot(2, 3, i)
    sns.countplot(data=df, x=var, hue="Emprestimo_pessoal", palette=cores)
    plt.title(f"Quantidades de {var}")
    plt.legend(title="Emprestimo", labels=["NÃ£o Aceitou", "Aceitou"])

# TÃ­tulo geral do grÃ¡fico
plt.suptitle("AnÃ¡lise em relaÃ§Ã£o a Emprestimo pessoal", fontsize=16)
plt.show()
```

## limpezas e Tratamento dos Dados

```python
df.columns
```

```python
# Lista das colunas nÃ£o relevantes
colunas_nao_relevantes = ['ID', 'ZIP_code']

# Excluir as colunas nÃ£o relevantes
df = df.drop(colunas_nao_relevantes, axis=1)

# Exibir as primeiras linhas do DataFrame apÃ³s a exclusÃ£o
df.head()
```

```python
# Calcular a matriz de correlaÃ§Ã£o
correlation_matrix = df.corr()

# Configurar o tamanho da figura
plt.figure(figsize=(10, 8))

# MÃ¡scara para a parte superior da matriz de correlaÃ§Ã£o
# mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Criar um mapa de calor (heatmap) da matriz de correlaÃ§Ã£o
sns.heatmap(correlation_matrix, annot=True, cmap='Greens', fmt=".2f")

# TÃ­tulo do grÃ¡fico
plt.title("Matriz de CorrelaÃ§Ã£o entre VariÃ¡veis")

# Exibir o grÃ¡fico
plt.show()
```

```python
df = df.drop('Experiencia', axis=1)
```

```python
df.head()
```

```python
# Configurar o tamanho da figura com 1 linha e 2 colunas
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Criar grÃ¡ficos de caixa para cada metade das variÃ¡veis
sns.boxplot(data=df.iloc[:, :5], ax=axes[0], orient="v", palette="Set3")
sns.boxplot(data=df.iloc[:, 5:], ax=axes[1], orient="v", palette="Set3")

# TÃ­tulos dos grÃ¡ficos
axes[0].set_title("GrÃ¡fico de Caixa (Parte 1)")
axes[1].set_title("GrÃ¡fico de Caixa (Parte 2)")

# Ajustar o layout dos grÃ¡ficos
plt.tight_layout()

# Exibir os grÃ¡ficos
plt.show()
```

```python
from scipy import stats

z_scores = stats.zscore(df['Hipoteca'])
limite = 3
outliers = df['Hipoteca'][abs(z_scores) > limite]
df = df.drop(outliers.index)
```

```python
# Configurar o tamanho da figura com 1 linha e 2 colunas
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Criar grÃ¡ficos de caixa para cada metade das variÃ¡veis
sns.boxplot(data=df.iloc[:, :5], ax=axes[0], orient="v", palette="Set3")
sns.boxplot(data=df.iloc[:, 5:], ax=axes[1], orient="v", palette="Set3")

# TÃ­tulos dos grÃ¡ficos
axes[0].set_title("GrÃ¡fico de Caixa (Parte 1)")
axes[1].set_title("GrÃ¡fico de Caixa (Parte 2)")

# Ajustar o layout dos grÃ¡ficos
plt.tight_layout()

# Exibir os grÃ¡ficos
plt.show()
```
