```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, confusion_matrix


# Inicializar o DummyClassifier como um modelo baseline
dummy_classifier = DummyClassifier(strategy="most_frequent")  # EstratÃ©gia de previsÃ£o mais frequente

# Treinar o modelo baseline
dummy_classifier.fit(X_train, y_train)

# Fazer previsÃµes no conjunto de teste
y_pred = dummy_classifier.predict(X_test)

# Calcular a acurÃ¡cia do modelo baseline
accuracy = accuracy_score(y_test, y_pred)
print("AcurÃ¡cia do modelo baseline:", accuracy)

# Calcular a matriz de confusÃ£o
cm = confusion_matrix(y_test, y_pred, labels=dummy_classifier.classes_)

# Exibir a matriz de confusÃ£o com ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dummy_classifier.classes_)
disp.plot(cmap=plt.cm.Blues, values_format="d")
plt.title("Matriz de ConfusÃ£o - Modelo Baseline")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Inicializar e treinar o Dummy Classifier
dummy_classifier = DummyClassifier(strategy="most_frequent")
dummy_classifier.fit(X_train, y_train)
y_pred_dummy = dummy_classifier.predict(X_test)
accuracy_dummy = accuracy_score(y_test, y_pred_dummy)
print(f"AcurÃ¡cia do Dummy Classifier: {accuracy_dummy:.4f} %")

# Normalizar as features numÃ©ricas usando MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Inicializar e treinar a Logistic Regression
logreg = LogisticRegression(random_state=42)
logreg.fit(X_train_scaled, y_train)
y_pred_logreg = logreg.predict(X_test_scaled)
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print(f"AcurÃ¡cia da Logistic Regression: {accuracy_logreg:.4f} %")

# Inicializar e treinar a Decision Tree
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train_scaled, y_train)
y_pred_dt = decision_tree.predict(X_test_scaled)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"AcurÃ¡cia da Decision Tree: {accuracy_dt:.4f} %")
```

```python
# FunÃ§Ã£o para plotar a matriz de confusÃ£o
def plot_confusion_matrix(model, model_name, X_test, y_test):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap=plt.cm.Blues, values_format="d")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

# Chamar a funÃ§Ã£o para cada modelo
plot_confusion_matrix(dummy_classifier, "Dummy Classifier", X_test_scaled, y_test)
plot_confusion_matrix(logreg, "Logistic Regression", X_test_scaled, y_test)
plot_confusion_matrix(decision_tree, "Decision Tree", X_test_scaled, y_test)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# FunÃ§Ã£o para plotar a curva ROC
def plot_roc_curve(y_true, y_scores, label):
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    auc_score = roc_auc_score(y_true, y_scores)
    plt.plot(fpr, tpr, label=f'{label} (AUC = {auc_score:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')  # Linha de referÃªncia para um classificador aleatÃ³rio
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()

# Obtendo as probabilidades das classes positivas
dummy_probs = dummy_classifier.predict_proba(X_test)[:, 1]
logreg_probs = logreg.predict_proba(X_test)[:, 1]
dt_probs = decision_tree.predict_proba(X_test)[:, 1]

# Plotando a curva ROC para cada modelo
plot_roc_curve(y_test, dummy_probs, label='Dummy Classifier')
plot_roc_curve(y_test, logreg_probs, label='Logistic Regression')
plot_roc_curve(y_test, dt_probs, label='Decision Tree')

# Exibindo o grÃ¡fico
plt.show()
```

```python
""" 
Escreva o cÃ³digo para realizar a validaÃ§Ã£o cruzada nos modelos "logreg_model" e "dt_model",
com a classe "StratifiedKFold" e a funÃ§Ã£o "cross_validate".

Siga os seguintes passos:
1. Defina o nÃºmero de divisÃµes e a estratÃ©gia de embaralhamento
2. Defina um random state
3. Realize a validaÃ§Ã£o cruzada para cada modelo
4. Obtenha os valores de AUC para cada divisÃ£o para os dados de treino e teste
5. Calcule a mÃ©dia e o desvio-padrÃ£o do AUC para os dados de treino e teste
6. Exiba os resultados da validaÃ§Ã£o cruzada
"""
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate

# Inicializar os modelos
logreg_model = LogisticRegression(random_state=42)
dt_model = DecisionTreeClassifier(random_state=42)

# Definir o nÃºmero de divisÃµes e a estratÃ©gia de embaralhamento
n_splits = 10

# Definir um random state
seed = 42

# Inicializar o objeto StratifiedKFold
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)

# Realizar a validaÃ§Ã£o cruzada para cada modelo
logreg_scores = cross_validate(logreg_model, X, y, cv=skf, scoring='roc_auc', return_train_score=True)
dt_scores = cross_validate(dt_model, X, y, cv=skf, scoring='roc_auc', return_train_score=True)

# FunÃ§Ã£o para exibir os resultados da validaÃ§Ã£o cruzada
def display_scores(scores, model_name):
    train_auc_mean = scores['train_score'].mean()
    train_auc_std = scores['train_score'].std()
    test_auc_mean = scores['test_score'].mean()
    test_auc_std = scores['test_score'].std()

    print(f"Modelo: {model_name}")
    print(f"Treino - AUC mÃ©dio: {train_auc_mean:.4f} (Desvio-padrÃ£o: {train_auc_std:.4f})")
    print(f"Teste - AUC mÃ©dio: {test_auc_mean:.4f} (Desvio-padrÃ£o: {test_auc_std:4f})")
    print()

# Exibir os resultados da validaÃ§Ã£o cruzada para cada modelo
display_scores(logreg_scores, "Logistic Regression")
display_scores(dt_scores, "Decision Tree")
```

```python
def validacao_cruzada(model, model_name):
    from sklearn.model_selection import cross_val_score, StratifiedKFold, cross_validate
    
    splits = 10
    seed = 42
    stk = StratifiedKFold(n_splits=splits, shuffle=True, random_state=seed)
    
    cross = cross_validate(estimator=model, X=X, y=y, scoring='roc_auc', cv=stk, return_train_score=True)
    train_score = cross['train_score'].mean()
    train_score_std = cross['train_score'].std()
    
    test_score = cross['test_score'].mean()
    test_score_std = cross['test_score'].std()
    
    print(f"Modelo: {model_name}")
    print(f"Treino - AUC mÃ©dio: {train_score:.4f} (Desvio-padrÃ£o: {train_score_std:.4f})")
    print(f"Teste - AUC mÃ©dio: {test_score:.4f} (Desvio-padrÃ£o: {test_score_std:4f})")
    print()
    
validacao_cruzada(random_forest, model_name='Random Forest')
validacao_cruzada(grad_boost, model_name='Gradient Bossting')
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold, cross_validate
from sklearn.metrics import make_scorer, roc_auc_score

# Definir a quantidade de divisÃµes e repetiÃ§Ãµes
n_splits = 5  # NÃºmero de divisÃµes (folds)
n_repeats = 10  # NÃºmero de repetiÃ§Ãµes

# Criar o objeto RepeatedStratifiedKFold
rskf = RepeatedStratifiedKFold(n_splits=n_splits, n_repeats=n_repeats, random_state=42)

# Definir a mÃ©trica para avaliaÃ§Ã£o (AUC)
scoring = {'auc': make_scorer(roc_auc_score, needs_proba=True)}

# Realizar a validaÃ§Ã£o cruzada para Logistic Regression
logreg_scores = cross_validate(logreg_model_balanced, X_train_balanced, y_train_balanced, cv=rskf, scoring=scoring, return_train_score=True)

# Realizar a validaÃ§Ã£o cruzada para Decision Tree
dt_scores = cross_validate(dt_model_balanced, X_train_balanced, y_train_balanced, cv=rskf, scoring=scoring, return_train_score=True)

# Calcular os valores mÃ©dios de AUC e desvio padrÃ£o
mean_auc_logreg_train = logreg_scores['train_auc'].mean()
mean_auc_logreg_test = logreg_scores['test_auc'].mean()
std_auc_logreg_test = logreg_scores['test_auc'].std()

mean_auc_dt_train = dt_scores['train_auc'].mean()
mean_auc_dt_test = dt_scores['test_auc'].mean()
std_auc_dt_test = dt_scores['test_auc'].std()

# Exibir os resultados
print("Logistic Regression - AUC (Treino):", mean_auc_logreg_train)
print("Logistic Regression - AUC (Teste):", mean_auc_logreg_test)
print("Logistic Regression - Desvio PadrÃ£o AUC (Teste):", std_auc_logreg_test)

print("Decision Tree - AUC (Treino):", mean_auc_dt_train)
print("Decision Tree - AUC (Teste):", mean_auc_dt_test)
print("Decision Tree - Desvio PadrÃ£o AUC (Teste):", std_auc_dt_test)
```

## Ajustando os hiperparÃ¢metros na Arvore de DecisÃ£o

```python
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score

# Definir os parÃ¢metros que deseja otimizar
param_grid = {
    'max_depth': [None, 10, 20, 30],  # Profundidade mÃ¡xima da Ã¡rvore
    'min_samples_split': [2, 5, 10],  # NÃºmero mÃ­nimo de amostras para dividir um nÃ³
    'min_samples_leaf': [1, 2, 4],  # NÃºmero mÃ­nimo de amostras em uma folha
    'criterion' : ["gini", "entropy", "log_loss"]
}

# Crie o modelo Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)

# Crie o objeto GridSearchCV
grid_search = GridSearchCV(dt_model, param_grid=param_grid, cv=10, scoring='roc_auc')

# Ajuste o modelo aos dados de treino usando a busca em grade
grid_search.fit(X_train, y_train)

# Exiba os melhores parÃ¢metros encontrados
best_params = grid_search.best_params_
print("Melhores parÃ¢metros:", best_params)

# Calcule o AUC nos dados de treino
y_train_pred = grid_search.predict_proba(X_train)[:, 1]
auc_train = roc_auc_score(y_train, y_train_pred)
print("AUC nos dados de treino:", auc_train)

# Calcule o AUC nos dados de teste
y_test_pred = grid_search.predict_proba(X_test)[:, 1]
auc_test = roc_auc_score(y_test, y_test_pred)
print("AUC nos dados de teste:", auc_test)

dt_model_best = grid_search.best_estimator_
dt_preds = dt_model_best.predict(X_test)

# Classification Report
print('---'*20)
print(classification_report(y_test, dt_preds))
```
